# Create a Separate Dataframe for Most Figures  
This dataframe will be tidy data formatted with years and time frames

```python
Tidy2010Subject = pd.read_csv('../../data/Tidy2010Subject.csv')
Tidy2020Subject = pd.read_csv('../../data/Tidy2020Subject.csv')
Tidy2000SF1 = pd.read_csv('../../data/Tidy2000SF1.csv')
Tidy2000SF3 = pd.read_csv('../../data/Tidy2000SF3.csv')
Tidy2010SF1 = pd.read_csv('../../data/Tidy2010SF1.csv')
Tidy2010ACS = pd.read_csv('../../data/Tidy2010ACS.csv')
Tidy2020PL = pd.read_csv('../../data/Tidy2020PL.csv')
Tidy2020ACS = pd.read_csv('../../data/Tidy2020ACS.csv')
```

```python
dfs2000 = [Tidy2000SF1, Tidy2000SF3]
dfs2010 = [Tidy2010SF1, Tidy2010ACS, Tidy2010Subject]
dfs2020 = [Tidy2020PL, Tidy2020ACS, Tidy2020Subject]
#merging 2000's dataframes and adding multi-index for both year and time frame categories
twothousand = Tidy2000SF1.merge(Tidy2000SF3, on = 'NAME')
twothousand['Year'] = 2000
twothousand = twothousand.set_index(['NAME'])
twothousand.columns = pd.MultiIndex.from_product([twothousand.columns, ['2000'], ['None']])
#merging 2010's dataframes and adding multi-index for both year and time frame categories
twothousandten = reduce(lambda  left,right: pd.merge(left,right,on=['NAME'],
                                                     how='outer'), dfs2010)
twothousandten['Year'] = 2010
twothousandten = twothousandten.set_index(['NAME'])
twothousandten.columns = pd.MultiIndex.from_product([twothousandten.columns, ['2010'], ['None']])
#merging 2020's dataframes and adding multi-index for both year and time frame categories
twothousandtwenty = reduce(lambda  left,right: pd.merge(left,right,on=['NAME'],
                                                        how='outer'), dfs2020)
twothousandtwenty['Year'] = 2020
twothousandtwenty = twothousandtwenty.set_index(['NAME'])
twothousandtwenty.columns = pd.MultiIndex.from_product([twothousandtwenty.columns, ['2020'], ['None']])
dfs = [twothousand, twothousandten, twothousandtwenty]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['NAME'],
                                            how='outer'), dfs)
df = df_merged
df.head(2)
```

```python
#make a list of all columns to run through for calculating percent change and real change over all time periods
columns = list(df.columns)
columns = columns[:]
element = []
for x in columns:
    element.append(x[0])
```

```python
#loop through elements
for i in element:
    df[('{} % Change'.format(i), 'None','2000-2020')] = percentchange(df[('{}'.format(i),'2020','None')], df[('{}'.format(i),'2000','None')])
    df[('{} % Change'.format(i), 'None','2000-2010')] = percentchange(df[('{}'.format(i),'2010','None')], df[('{}'.format(i),'2000','None')])
    df[('{} % Change'.format(i), 'None','2010-2020')] = percentchange(df[('{}'.format(i),'2020','None')], df[('{}'.format(i),'2010','None')])
    df[('{} Change'.format(i), 'None','2000-2020')] = (df[('{}'.format(i),'2020','None')] - df[('{}'.format(i),'2000','None')])
    df[('{} Change'.format(i), 'None','2000-2010')] = (df[('{}'.format(i),'2010','None')] - df[('{}'.format(i),'2000','None')])
    df[('{} Change'.format(i), 'None','2010-2020')] = (df[('{}'.format(i),'2020','None')] - df[('{}'.format(i),'2010','None')])
```

```python
#stack and rename to get tidy data for both year and time period - half of dataframe will be null depending on if it's a change or point in time metric
df = df.stack([1, 1])
df = df.reset_index(drop = False)
df = df.rename(columns = {'level_1':'Year', 'level_2':'Time Frame'})
```
