```python
def prettypercentchange(x, y):
    return round((x - y)*100/y, 1).astype(str)+'%'
def mathpercentchange(x, y):
    return round((x - y)/y, 1)
```

```python
Prefix2010Subject = pd.read_csv('../../data/Prefix2010Subject.csv')
Prefix2020Subject = pd.read_csv('../../data/Prefix2020Subject.csv')
Prefix2000SF1 = pd.read_csv('../../data/Prefix2000SF1.csv')
Prefix2000SF3 = pd.read_csv('../../data/Prefix2000SF3.csv')
Prefix2010SF1 = pd.read_csv('../../data/Prefix2010SF1.csv')
Prefix2010ACS = pd.read_csv('../../data/Prefix2010ACS.csv')
Prefix2020PL = pd.read_csv('../../data/Prefix2020PL.csv')
Prefix2020ACS = pd.read_csv('../../data/Prefix2020ACS.csv')
```

```python
dfs = [Prefix2000SF1, Prefix2000SF3, Prefix2010SF1, Prefix2010ACS, Prefix2010Subject, Prefix2020PL, Prefix2020ACS, Prefix2020Subject]
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['NAME'],
                                            how='outer'), dfs)
```

```python
transp = df_merged.transpose()
transp.columns = transp.iloc[0]
transp = transp.loc[transp['Stewart County, Tennessee'] != 'Stewart County, Tennessee']
```
