```python
folium.Map(location=[41.919950, -87.708990], tiles = 'CartoDB positron')
```

```python
#This shows how you can add layer control
tiles_map=folium.Map(location=[41.919950, -87.708990])
folium.TileLayer('Stamen Terrain').add_to(tiles_map)
folium.TileLayer('Stamen Toner').add_to(tiles_map)
folium.TileLayer('Stamen Water Color').add_to(tiles_map)
folium.TileLayer('cartodbpositron').add_to(tiles_map)
folium.TileLayer('cartodbdark_matter').add_to(tiles_map)
folium.LayerControl().add_to(tiles_map)
tiles_map
```

```python
#zoom_start sets initial zoom level, default is at 10 the higher the number the more zoomed you are
folium.Map(location=[41.919950, -87.708990], tiles = 'CartoDB positron', zoom_start = 15)
```

Markers  
There are many marker types, with the default being the little blue drop.

```python
m = folium.Map(location=[41.919950, -87.708990], tiles = 'CartoDB positron', zoom_start = 15)

tooltip = "Click for More Information"

folium.Marker([41.919950, -87.708990], popup = "<b>Joanna's Old Coach House</b>", tooltip = tooltip).add_to(m)

folium.Marker([41.917385, -87.707809], popup = '<b>Go Tavern</b>', tooltip = tooltip).add_to(m)

folium.Marker([41.921352, -87.703826], popup = '<b>RIP Crooked Tree</b>', tooltip = tooltip).add_to(m)

m
```

Built in support for colors and marker icon types from bootstrap.  
Use FontAwesome 4 icons, find 'em here: https://fontawesome.com/v4/icons/

```python
m = folium.Map(location=[41.919950, -87.708990], tiles = 'CartoDB positron', zoom_start = 15)

folium.Marker([41.919950, -87.708990], tooltip = "<b>Joanna's Old Coach House</b>", icon = folium.Icon(color = 'pink', icon = 'sun', prefix = 'fa')).add_to(m)
#circle in meters
folium.Circle(radius=100, location=[41.919950, -87.708990], popup="The Walk Zone", color="crimson", fill=False).add_to(m)

folium.Marker([41.917385, -87.707809], tooltip = '<b>Go Tavern</b>', icon=folium.Icon(color="blue", icon = 'trophy', prefix = 'fa')).add_to(m)

folium.Marker([41.921352, -87.703826], tooltip = '<b>RIP Crooked Tree</b>', icon = folium.Icon(color = 'green', icon = 'tree', prefix = 'fa')).add_to(m)
#circle in pixels
folium.CircleMarker(location=[41.921352, -87.703826], radius=50, tooltip="Crooked Tree Zone", color="#3186cc", fill=True, fill_color="#3186cc").add_to(m)

m
```
